```python
upskill = []
placeholder1=''
sag=0
for i in rows:
    #fields = [k.text.strip() for k in i.find_all('td')][1:-1]
    fields =i.find('a')['href']
    link='https:'+fields
#     data.append(link)
    driver.get(link)
    soupn = bs(driver.page_source,'lxml')
    tit=' '.join(soupn.find('div',{'class':'col-md-7'}).find('h1').text.split())
    maio=' '.join(soupn.find('div',{'id':'topbar'}).find_all('div',{'class':'col-sm-6 col-md-5'})[-1].get_text(separator='\\').split()).split('\\')
    for va in maio:
        if '@' in va:
            mll=va.strip()
    #mla=(va.strip() for va in maio if '@' in va)
    mai=soupn.find('section',{'class':'inner-page'}).find('div',{'class':'col-md-7'}).find_all('p',{'class':'MsoNormal'})

    for idx,a in enumerate(mai):
        if a.text=='Introduction':
            info=mai[idx+1].text
        elif 'INTRODUCTION' in a.text:
            info=mai[idx].text.replace('INTRODUCTION','').strip()
    #mll
    if mll=='':
        mal_=''
    else:
        mal_=json.dumps([mll])
    tab=soupn.find('section',{'class':'inner-page'}).find('div',attrs={'class':'col-md-5'}).find('table')
    shi=tab.tbody.find_all('tr')
    sar=list(' '.join(a.text.split()) for a in shi)
    
    swi=[re.findall('\d{2}/\d{2}/\d{4}',v) for v in sar]
    smith=[re.search('\$\W?\d*\W?\d+',v).group().replace(',','').split() for v in sar]
    
    
    for idx,ases in enumerate(swi):
        spl_dt_obj = [datetime.strptime(am, '%d/%m/%Y') for am in ases]
        ate_= [z.strftime('%Y-%m-%d') for z in spl_dt_obj]
        start_date=ate_[0]
        end_date=ate_[1]
        title=tit
        event_info=info
        mail_=mal_
        swp=smith[idx]
        
        if int(swp[1])>0:
            ty='paid'
            cu=swp[0]
            pr=swp[1]
        else:
            ty='Free'
            cu=''
            pr=''
        if ty=='':
            ticket_list=''
        else:
            ticketlist={'type': ty,
                         'price':pr,
                         'currency': cu}
            ticket_list=json.dumps([ticketlist])
            
            
        ####################################
        def get_google_map_url(location):
            try:
                google_url_for_location="https://www.google.com/search?q="+location+"&oq="+location+"&num=1"
                time.sleep(randint(0,3))
                driver.get(google_url_for_location)
                time.sleep(4)
                try:
                    google_map_url=driver.find_element_by_id('lu_map').click()
                except:
                    try:
                        google_map_url=driver.find_element_by_class_name('Xm7sWb').click()
                    except:
                        google_map_url=driver.find_element_by_class_name('Lx2b0d').click()
                time.sleep(1)
                google_map_url=driver.current_url
    #                 print(google_map_url)
                return(google_map_url)
            except Exception as e:
                print(location, "; didn't go through")
                print(e)
                return("")
        ##################
        try:    
            we=soupn.find('section',{'class':'inner-page'}).find('div',attrs={'class':'col-md-5'}).h2.text
            pe=re.search('\(.*\)',we).group().replace('(','').replace(')','').split(',')
            city= pe[0].strip()
            country=pe[1].strip()
            on_off='0'
            venue=pe[0].strip()
            #googlePlaceUrl=get_google_map_url(venue)
            
            if venue==placeholder1:
                googlePlaceUrl=googlePlaceUrl
            else:
                googlePlaceUrl=get_google_map_url(venue)
            placeholder1=pe[0].strip()
        
        except:
            on_off='1'
            city=''
            country=''
            venue=''
            googlePlaceUrl=''
        org_pro="""Upskill Development Institute is a training, research and consultancy services provider to international organizations. We assist our clients to improve their skills and confident by offering the best possible solutions to enable them fulfil their objectives for a sustainable result. We strive to improve the quality and accessibility of the training by creating an environment that ensures successful learning from start to finish."""
        org_web='https://upskilldevelopment.com/'
        org_name='Upskill Development Institute'
        
        logo=''
        sponsors=''
        agendalist=''
        type_=''
        category=''
        Speakerlist=''
        event_web=link
        time_=''
       
        event_web=link
    
        
        
        upskill.append([link,title,start_date,end_date,time_,event_info,ticket_list,
                 org_pro,org_name,org_web,logo,sponsors,agendalist,
                 type_,category,city,country,venue,event_web,googlePlaceUrl,mail_,
                 Speakerlist,on_off])
    sag+=1
    for d in range(0,50*100,100):
        if sag==d:
            print(d,' cards done')
print('Done')
```
