```python
n=1
brek=True
cards=[]
while brek:
    url=f'https://events.umich.edu/list?page={n}'
    driver.get(url)
    time.sleep(3)
    soup=bs(driver.page_source,'lxml')
    car=soup.find_all('div',{'class':'event-listing-grid'})
    
    for j in car:
        ritle=' '.join(j.find('div',{'class':'event-title'}).text.split())
        web=j.find('a')['href']
        link='https://events.umich.edu'+web
        dat_dur=' '.join(j.find('time',{'class':'time-banner'}).text.split())
        dat=dat_dur.replace('–','-')
        if '(' in dat:
            dat=dat.split('(')[0].strip()
        if dat=='':
            start_date=end_date=''
        else:
            mad=dat
            #try:
    #             ma=re.search('([A-Sa-z]+\W+\d{1,2}.+\W+\d{4}|\d{1,2}.+\W+\d{4})',_date)#, maxsplit=0)
    #             mad=ma.group()

            if '-' in mad:
                st=mad.split('-')[0].strip().replace(',','').replace('nd','').replace('st','').replace('th','').replace('rd','').replace('ugu','ugust')
                en=mad.split('-')[1].strip().replace(',','').replace('nd','').replace('st','').replace('th','').replace('rd','').replace('ugu','ugust')

            elif 'and'in mad:
                st=mad.split('and')[0].strip().replace(',','').replace('nd','').replace('st','').replace('th','').replace('rd','').replace('ugu','ugust')
                en=mad.split('and')[1].strip().replace(',','').replace('nd','').replace('st','').replace('th','').replace('rd','').replace('ugu','ugust')

            elif '&' in mad:
                st=mad.split('&')[0].strip().replace(',','').replace('nd','').replace('st','').replace('th','').replace('rd','').replace('ugu','ugust')
                en=mad.split('&')[1].strip().replace(',','').replace('nd','').replace('st','').replace('th','').replace('rd','').replace('ugu','ugust')

            else:
                st=mad.strip().replace(',','').replace('nd','').replace('st','').replace('th','').replace('rd','').replace('ugu','ugust')
                en=mad.strip().replace(',','').replace('nd','').replace('st','').replace('th','').replace('rd','').replace('ugu','ugust')
            #print(date__,st,en)
            ####################
            if any(c.isalpha() for c in st)==False:
            #      print('leg')
                pa=re.search('[A-Sa-z]+\W+(\d{4})',en)#, maxsplit=0)
                sapa=pa.group()
                start=st+' '+sapa
                end=en
            #                        02        April      2022
            elif re.search('\d{1,2}\s+[A-Sa-z]{3,9}\s\d{4}',st):
             #   print('awujale')
                start=st
                end=en

            #*                        April       02      2022
            elif re.search('[A-Sa-z]{3,9}\s\d{1,2}\s+\d{4}',st):
            #       print('ala')
                pa=re.search('(\d{4})',st).group()
                sa=re.search('([A-Sa-z]{3,9})',st).group()
                ta=re.search('(\d{1,2})',st).group()
                start=ta+' '+sa+' '+pa
                #
                ba=re.search('(\d{4})',en).group()
                ca=re.search('([A-Sa-z]{3,9})',en).group()
                da=re.search('(\d{1,2})',en).group()
                end=da+' '+ca+' '+ba     

            #                      02     2022
            elif re.search('\d{1,2}\s+\d{4}',en):
            #      print('kum')
                pa=re.search('(\d{4})',en).group()#, maxsplit=0)
                sa=re.search('([A-Sa-z]{3,9})',st).group()
                ta=re.search('(\d{1,2})',st).group()

                start=ta+' '+sa+' '+pa
                #
                ba=re.search('(\d{4})',en).group()
                ca=re.search('([A-Sa-z]{3,9})',st).group()
                da=re.search('(\d{1,2})',en).group()
                end=da+' '+ca+' '+ba

            #                      02         April
            elif re.search('\d{1,2}\s+[A-Sa-z]{3,9}',st):
            #        print('is')
                pa=re.search('(\d{4})',en)#, maxsplit=0)
                sapa=pa.group()
                start=st+' '+sapa
                end=en

            #*                      April           02
            elif re.search('[A-Sa-z]{3,9}\s+\d{1,2}',st):
            #      print('bad')
                pa=re.search('(\d{4})',en).group()#, maxsplit=0)
                sa=re.search('([A-Sa-z]{3,9})',st).group()
                ta=re.search('(\d{1,2})',st).group()

                start=ta+' '+sa+' '+pa
                #
                ba=re.search('(\d{4})',en).group()
                ca=re.search('([A-Sa-z]{3,9})',en).group()
                da=re.search('(\d{1,2})',en).group()
                end=da+' '+ca+' '+ba

            else:
            #     print('shik')
                start=end=''
            if start=='':
                start_date=end_date=''
            else:
                pick=[start, end]
                try:
                    spl_dt_obj = [datetime.strptime(v, '%d %B %Y') for v in pick]
                except:
                    spl_dt_obj = [datetime.strptime(v, '%d %b %Y') for v in pick]
                date_= [z.strftime('%Y-%m-%d') for z in spl_dt_obj]
                start_date=date_[0]
                end_date=date_[1]
        cards.append([link,ritle,start_date,end_date])
    n+=1
    if n==int(2)+1:#lastpage
        brek=False
#crads=(list(itertools.chain.from_iterable(cards)))
print('The total number of items to be scraped is ',len(cards))
```

```python
umich=[]

for i in cards:
    link=i[0]
    ritle=i[1]
    start_date=i[2]
    end_date=i[3]


    driver.get(link)
    soupn=bs(driver.page_source,'lxml')
    title=' '.join(soupn.find('h1',{'class':'title'}).text.split())
    event_info=' '.join(soupn.find('div',{'class':'event-description-text'}).text.split())
    org_name=' '.join(soupn.find('div',{'class':'col-sm-12'}).find('p').find('a').text.split())
    if org_name=='' or org_name==' ' or org_name=='-':
        org_name='University of Michigan'
    try:
        sha=soupn.find('div',{'class':'cosponsors'}).find_all('li')
        spons=[' '.join(a.text.split()) for a in sha]
    except:
        spons=''
    if spons=='' or spons==' ' or spons==[]:
        sponsor=''
    else:
        sponsor=json.dumps(spons)
    tim=' '.join(soupn.find('span',{'class':'time'}).text.split()).upper().split('-')
    if len(tim)==2:
        stt=tim[0].strip()
        ent=tim[1].strip()
        z=''
    elif len(tim)==1:
        stt=tim[0].strip()
        ent=''
        z=''
    time_st={
        'type':'general',
        'Start_time':stt,
        'end_time':ent,
        'timezone':z,
        'days':'all'
    }

    time_=json.dumps([time_st])
    #**************
    def gu(luc):
        google_url_for_location="https://www.google.com/search?q="+luc+"&oq="+luc+"&num=1"
        time.sleep(randint(0,3))
        driver.get(google_url_for_location)
        time.sleep(4)
        try:
            google_map_url=driver.find_element_by_id('lu_map').click()
        except:
            try:
                google_map_url=driver.find_element_by_class_name('Xm7sWb').click()
            except:
                google_map_url=driver.find_element_by_class_name('Lx2b0d').click()
        time.sleep(1)
        google_map_url=driver.current_url
#                 print(google_map_url)
        return(google_map_url)
        ######################################
    def get_google_map_url(location):
        try:
            return(gu(location))
        except:
            try:
                return(gu(location+', University of Michigan'))
            except:
                sha=location.split(',')
                try:
                    gu(sha[-3])
                except:
                    try:
                        gu(sha[-2])
                    except:
                        try:
                            gu(sha[-1])
                        except Exception as e:
                            print(location, "; url didn't go through")
                            print(e)
                            return("")


    def countr(locale):
        try:
            google_url_for_location="https://www.google.com/search?q="+locale+"&oq="+locale+"&num=1"
            time.sleep(randint(0,3))
            driver.get(google_url_for_location)
            time.sleep(4)
            try:
                google_map_url=driver.find_element_by_id('lu_map').click()
            except:
                try:
                    google_map_url=driver.find_element_by_class_name('Xm7sWb').click()
                except:
                    try:
                        google_map_url=driver.find_element_by_class_name('Lx2b0d').click()
                    except:
                        locale=locale+', University of Michigan'
                        google_url_for_location="https://www.google.com/search?q="+locale+"&oq="+locale+"&num=1"
                        time.sleep(randint(0,3))
                        driver.get(google_url_for_location)
                        time.sleep(4)
                        google_map_url=driver.find_element_by_id('lu_map').click()
            time.sleep(1)
            try:
                country=driver.find_element_by_class_name('x3AX1-LfntMc-header-title-VdSJob').text
            except:
                country=driver.find_element_by_class_name('bwoZTb').text
#                 print(country)
            return(country)
        except Exception as e:
            print(locale, ";country didn't go through")
            print(e)
            return("")

    try:
        loc=' '.join(soupn.find('span',{'class':'loc'}).text.split())
    except:
        loc=' '.join(soupn.find('a',{'class':'loc'}).text.split())
    ticket_list=''
    #org_name='The Linux Foundation®'
    org_web='https://events.umich.edu/'
    org_pro='Welcome to the University of Michigan Events Calendar! Use this site to browse and search for events happening throughout campus. The calendar system is full of events sponsored by registered student organizations and University departments. If you belong to a University department or registered student organization, you can add events to the calendar.'
    logo=''    
    #sponsor=''
    agendalist=''
    type_=''
    category=''
    Speakerlist='' 
    #time_=''
    event_web=link
    mail_=''

    if 'nline' in loc or 'ff Campus' in loc or loc=='' or loc==' ' or loc=='-' or loc=='Virtual' or 'nspecified' in loc:
        on_off='1'
        city=''
        venue=''
        country=''
        googlePlaceUrl=''
    else:
        on_off='0'
        if ',' in loc:
            city=loc.split(',')[-2].strip()
            country=loc.split(',')[-1].strip()
        else:
            city=loc
            country=countr(city)
        venue=loc
        googlePlaceUrl=get_google_map_url(venue)



    umich.append([link,title,start_date,end_date,time_,event_info,ticket_list,
                 org_pro,org_name,org_web,logo,sponsor,agendalist,
                 type_,category,city,country,venue,event_web,googlePlaceUrl,mail_,
                 Speakerlist,on_off])

    
print('Done')
```
