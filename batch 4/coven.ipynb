```python
url='https://kongresskalender.conventus.de/index.php?id=160&L=0'
#https://kongresskalender.conventus.de/index.php?id=160&L=0

driver.get(url)
time.sleep(5)

cards=[]

while True:
    soup = bs(driver.page_source,'lxml')
    card=soup.find_all('div',{'class':'card mb-45'})
    cards.append(card)
    try:
        dim=soup.find('ul',{'class':'pagination d-inline-flex'}).find('li',{'class':'page-item last next'}).find('a')['href']
        real='https://kongresskalender.conventus.de/'+dim
        driver.get(real)
    except:
        break

cards=list(itertools.chain.from_iterable(cards))  
```

```python
coven=[]
for m in cards:
# for k in range(1,22):
#     driver.switch_to.window(driver.window_handles[k])
#     time.sleep(1)
#     soupn=bs(driver.page_source,'lxml')
    
    link=m.find('a')['href']
    if link.endswith('://'):
        pass
#     elif link=='https://www.akademie-der-dwg.de/veranstaltungen/kursuebersicht/':
#         pass
    else:
        title=m.find('b').text
        dat=' '.join(m.find('div',{'class':'card-header pt-1 pb-1'}).text.split()).split('-')
        loc=' '.join(m.find('div',{'class':'col-md-3 text-lg-right'}).text.split())
#         driver.get(link)
        if len(dat)==1:
            pick=[dat[0].strip(),dat[0].strip()]
        else:
            pick=[a.strip() for a in dat]
        spl_dt_obj = [datetime.strptime(v.strip(), '%d.%m.%Y') for v in pick]
        
        date_= [z.strftime('%Y-%m-%d') for z in spl_dt_obj]
        start_date=date_[0]
        end_date=date_[1]
        
        
        def gu(luc):
            google_url_for_location="https://www.google.com/search?q="+luc+"&oq="+luc+"&num=1"
            time.sleep(randint(0,3))
            driver.get(google_url_for_location)
            time.sleep(4)
            try:
                google_map_url=driver.find_element_by_id('lu_map').click()
            except:
                try:
                    google_map_url=driver.find_element_by_class_name('Xm7sWb').click()
                except:
                    google_map_url=driver.find_element_by_class_name('Lx2b0d').click()
            time.sleep(1)
            google_map_url=driver.current_url
    #                 print(google_map_url)
            return(google_map_url)
            ######################################
        def get_google_map_url(location):
            try:
                return(gu(location))
            except:
                sha=location.split(',')
                try:
                    gu(sha[-3])
                except:
                    try:
                        gu(sha[-2])
                    except:
                        try:
                            gu(sha[-1])
                        except Exception as e:
                            print(location, "; url didn't go through")
                            print(e)
                            return("")


        def countr(locale):
            try:
                google_url_for_location="https://www.google.com/search?q="+locale+"&oq="+locale+"&num=1"
                time.sleep(randint(0,3))
                driver.get(google_url_for_location)
                time.sleep(4)
                try:
                    google_map_url=driver.find_element_by_id('lu_map').click()
                except:
                    try:
                        google_map_url=driver.find_element_by_class_name('Xm7sWb').click()
                    except:
                        google_map_url=driver.find_element_by_class_name('Lx2b0d').click()
                time.sleep(1)
                try:
                    country=driver.find_element_by_class_name('x3AX1-LfntMc-header-title-VdSJob').text
                except:
                    country=driver.find_element_by_class_name('bwoZTb').text
    #                 print(country)
                return(country)
            except Exception as e:
                print(locale, ";country didn't go through")
                print(e)
                return("")
        ticket_list=''
        org_name='Conventus Congressmanagement & Marketing'
        org_web='https://www.conventus.de/'
        org_pro=''
        logo=''    
        sponsor=''
        agendalist=''
        type_=''
        event_info=''
        time_=''
        Speakerlist=''
        category=''
        mail_=json.dumps(['post@conventus.de'])
        event_web=link

        if 'nline' in loc or 'ff Campus' in loc or loc=='' or loc==' ' or loc=='-' or loc=='Virtual' or 'nspecified' in loc or 'irtual' in loc:
            on_off='1'
            city=''
            venue=''
            country=''
            googlePlaceUrl=''
        else:
            on_off='0'
            city=loc
            country=countr(city)
            venue=loc
            googlePlaceUrl=get_google_map_url(venue)




        coven.append([link,title,start_date,end_date,time_,event_info,ticket_list,
                     org_pro,org_name,org_web,logo,sponsor,agendalist,
                     type_,category,city,country,venue,event_web,googlePlaceUrl,mail_,
                     Speakerlist,on_off])
print('Done')
```
