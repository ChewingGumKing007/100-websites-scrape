```python
sma=[]
for j in data:
    if 'membership' in j:
        pass
    else:
        driver.get(j)
        links=j
        soup2 = bs(driver.page_source,'lxml')
        try:
            date__ = soup2.find('p',{'class':'no-trailer-for-small-up'}).text.replace('–','-')
        except:
            date__=''
        #############
        #_date=mad
        #*****************DATE********************
        if date__=='':
            start_date=end_date=''
        else:
            mad=date__
            try:
    #             ma=re.search('([A-Sa-z]+\W+\d{1,2}.+\W+\d{4}|\d{1,2}.+\W+\d{4})',_date)#, maxsplit=0)
    #             mad=ma.group()

                if '-' in mad:
                    st=mad.split('-')[0].strip().replace(',','').replace('nd','').replace('st','').replace('th','').replace('rd','').replace('ugu','ugust')
                    en=mad.split('-')[1].strip().replace(',','').replace('nd','').replace('st','').replace('th','').replace('rd','').replace('ugu','ugust')

                elif 'and'in mad:
                    st=mad.split('and')[0].strip().replace(',','').replace('nd','').replace('st','').replace('th','').replace('rd','').replace('ugu','ugust')
                    en=mad.split('and')[1].strip().replace(',','').replace('nd','').replace('st','').replace('th','').replace('rd','').replace('ugu','ugust')

                elif '&' in mad:
                    st=mad.split('&')[0].strip().replace(',','').replace('nd','').replace('st','').replace('th','').replace('rd','').replace('ugu','ugust')
                    en=mad.split('&')[1].strip().replace(',','').replace('nd','').replace('st','').replace('th','').replace('rd','').replace('ugu','ugust')

                else:
                    st=mad.strip().replace(',','').replace('nd','').replace('st','').replace('th','').replace('rd','').replace('ugu','ugust')
                    en=mad.strip().replace(',','').replace('nd','').replace('st','').replace('th','').replace('rd','').replace('ugu','ugust')
                #print(date__,st,en)
                ####################
                if any(c.isalpha() for c in st)==False:
                #      print('leg')
                    pa=re.search('[A-Sa-z]+\W+(\d{4})',en)#, maxsplit=0)
                    sapa=pa.group()
                    start=st+' '+sapa
                    end=en
                #                        02        April      2022
                elif re.search('\d{1,2}\s+[A-Sa-z]{3,9}\s\d{4}',st):
                 #   print('awujale')
                    start=st
                    end=en

                #*                        April       02      2022
                elif re.search('[A-Sa-z]{3,9}\s\d{1,2}\s+\d{4}',st):
                #       print('ala')
                    pa=re.search('(\d{4})',st).group()
                    sa=re.search('([A-Sa-z]{3,9})',st).group()
                    ta=re.search('(\d{1,2})',st).group()
                    start=ta+' '+sa+' '+pa
                    #
                    ba=re.search('(\d{4})',en).group()
                    ca=re.search('([A-Sa-z]{3,9})',en).group()
                    da=re.search('(\d{1,2})',en).group()
                    end=da+' '+ca+' '+ba     

                #                      02     2022
                elif re.search('\d{1,2}\s+\d{4}',en):
                #      print('kum')
                    pa=re.search('(\d{4})',en).group()#, maxsplit=0)
                    sa=re.search('([A-Sa-z]{3,9})',st).group()
                    ta=re.search('(\d{1,2})',st).group()

                    start=ta+' '+sa+' '+pa
                    #
                    ba=re.search('(\d{4})',en).group()
                    ca=re.search('([A-Sa-z]{3,9})',st).group()
                    da=re.search('(\d{1,2})',en).group()
                    end=da+' '+ca+' '+ba

                #                      02         April
                elif re.search('\d{1,2}\s+[A-Sa-z]{3,9}',st):
                #        print('is')
                    pa=re.search('(\d{4})',en)#, maxsplit=0)
                    sapa=pa.group()
                    start=st+' '+sapa
                    end=en

                #*                      April           02
                elif re.search('[A-Sa-z]{3,9}\s+\d{1,2}',st):
                #      print('bad')
                    pa=re.search('(\d{4})',en).group()#, maxsplit=0)
                    sa=re.search('([A-Sa-z]{3,9})',st).group()
                    ta=re.search('(\d{1,2})',st).group()

                    start=ta+' '+sa+' '+pa
                    #
                    ba=re.search('(\d{4})',en).group()
                    ca=re.search('([A-Sa-z]{3,9})',en).group()
                    da=re.search('(\d{1,2})',en).group()
                    end=da+' '+ca+' '+ba

                else:
                #     print('shik')
                    start=end=''
                if start=='':
                    start_date=end_date=''
                else:
                    pick=[start, end]
                    spl_dt_obj = [datetime.strptime(v, '%d %B %Y') for v in pick]
                    date_= [z.strftime('%Y-%m-%d') for z in spl_dt_obj]
                    start_date=date_[0]
                    end_date=date_[1]
            except:
                start_date=end_date=''





        ###############

    #     if '-' in _date:
    #         ash=_date.split('-')
    #         start_date=
    #     elif '&' in _date:
    #     else:


        topic_location = soup2.find('div',{'class':topic})
        title = topic_location.find('h1').text

        loc = topic_location.find_all('p')[-1].text
        if 'Online' in loc or 'Webinar' in loc:
            on_off='1'
            loc=''
        else:
            on_off='0'
            loc=loc
        #'Prices start from £90.00

        ct_ = soup2.find('p',{'class':cost})#.text.strip()#prices start from <?
        if 'Share' in str(ct_) or str(ct_)=='None':
            ticketlist=''
        else:
            cost_=str(ct_).replace('<strong','</strong').split('</strong>')[-2]
            if '£' in cost_:
                ala=cost_.split('£')
                if float(ala[1])>0:
                    ty='paid'
                    am=ala[1]
                    cu='£'

                else:
                    ty='free'
                    am=''
                    cu=''
            elif '$' in cost_:
                ala=cost_.split('$')
                if float(ala[1])>0:
                    ty='paid'
                    am=ala[1]
                    cu='$'
                else:
                    ty='free'
                    am=''
                    cu=''
            elif '€' in cost_:
                ala=cost_.split('€')
                if float(ala[1])>0:
                    ty='paid'
                    am=ala[1]
                    cu='€'
                else:
                    ty='free'
                    am=''
                    cu=''
            else:
                ty=''
                am=''
                cu=''
            if ty=='':
                ticketlist=''
            else:
                ticketlist={'type': ty,
                             'price':am,
                             'currency': cu}
           # print(ty,am,cu)
        try:    
            orm=topic_location.find_all('p')[-5].text
            ormi=topic_location.find_all('p')[-3].text
        except:
            orm=''
            ormi=''
        if 'Organi' in orm:
            try:
                org_na=topic_location.find_all('p')[-3].text
            except:
                org_na=''
            try:
                org_w=topic_location.find_all('p')[-3].find('a')['href']
                or_w='https://www.soci.org'+org_w
            except:
                or_w='https://www.soci.org/'#allley


        elif 'Organi' in ormi:
            try:
                org_na=topic_location.find_all('p')[-2].text
            except:
                org_na=''
            try:
                org_w=topic_location.find_all('p')[-2].find('a')['href']
                if 'http' in org_w:
                    or_w=org_w
                else:
                    or_w='https://www.soci.org'+org_w
            except:
                or_w='https://www.soci.org/'

        else:
            org_na=''
            or_w=''
        try:    
            ev_i=soup2.find('div',{'class':'event-content-panel'})
            ev=ev_i.find_all('p')
            if 'img' in str(ev[0]):
                even=' '.join(ev[1].text.split())
            else:
                even=' '.join(ev[0].text.split())

        except:
            even=''

        try:
            alo = soup2.find_all('div',{'class':agenda})#.find('dl')##needs sorting
            ag=[yup.find('dl') for yup in alo]
            a_ag = [qp.find_all() for qp in ag]
            all_ag=(list(itertools.chain.from_iterable(a_ag)))
            agenda_ = []
            temp = []
            for l in all_ag:

                if '<dt>' in str(l):
                    if len(temp) !=0:
                        agenda_.append([temp[0].text,[a.get_text(separator=', ') for a in temp[1:]]])
                        temp=[]
                        temp.append(l)
                    else:
                        temp.append(l)
                else:
                    temp.append(l)
        except:
            agenda_=[]
        #print(agenda_)
        rack=[]
        for idx,z in enumerate(agenda_):
            vtt=z[0].strip()
            if '.' in vtt:
                st=''
            elif vtt!='':
                spl = datetime.strptime(vtt, '%H:%M')
                st= spl.strftime('%I:%M%p')
            
            else:
                st=vtt
            if idx+1!=len(agenda_):
                wtt=agenda_[idx+1][0].strip()
                
                if '.' in wtt:
                    et=''
                elif wtt!='':
                    spl_ = datetime.strptime(wtt, '%H:%M')
                    et= spl_.strftime('%I:%M%p')
                else:
                    et=wtt
            else:
                et=''
            desc=''
            titl=z[1][0]
            da=start_date
            ori={'start_time':st,
                 'end_time':et,
                 'day':da,
                 'desc':desc,
                 'title':titl}
            rack.append(ori)

        try:
            spe = soup2.find('div',{'id':speakers}).find_all('div',{'class':'columns small-12 medium-9'})
            speak=[' '.join(s.text.split()) for s in spe]
        except:
            speak =''
        try:
            v_contact = soup2.find('div',{'class':venue_contact})
            v_contact_list = v_contact.find_all('p')
            address =' '.join(v_contact_list[1].text.split())
            #phone = v_contact_list[-2].text
            email = v_contact_list[-1].text.split(':').strip()
        except:
            address = ''
            #phone = ''
            email = 'conferences@soci.org'

        ###############COUNTYR,GPU#################
        def countr(locale):
            try:
                google_url_for_location="https://www.google.com/search?q="+locale+"&oq="+locale+"&num=1"
                time.sleep(randint(0,3))
                driver.get(google_url_for_location)
                time.sleep(4)
                try:
                    google_map_url=driver.find_element_by_id('lu_map').click()
                except:
                    try:
                        google_map_url=driver.find_element_by_class_name('Xm7sWb').click()
                    except:
                        google_map_url=driver.find_element_by_class_name('Lx2b0d').click()
                time.sleep(1)
                try:
                    country=driver.find_element_by_class_name('x3AX1-LfntMc-header-title-VdSJob').text
                except:
                    country=driver.find_element_by_class_name('bwoZTb').text
    #                 print(country)
                return(country)
            except Exception as e:
                print(locale, ";country didn't go through")
                print(e)
                return("")

        ####################################
        def gu(luc):
            google_url_for_location="https://www.google.com/search?q="+luc+"&oq="+luc+"&num=1"
            time.sleep(randint(0,3))
            driver.get(google_url_for_location)
            time.sleep(4)
            try:
                google_map_url=driver.find_element_by_id('lu_map').click()
            except:
                try:
                    google_map_url=driver.find_element_by_class_name('Xm7sWb').click()
                except:
                    google_map_url=driver.find_element_by_class_name('Lx2b0d').click()
            time.sleep(1)
            google_map_url=driver.current_url
#                 print(google_map_url)
            return(google_map_url)
            ######################################
        def get_google_map_url(location):
            try:
                return(gu(location))
            except:
                sha=location.split(',')
                try:
                    gu(sha[-3])
                except:
                    try:
                        gu(sha[-2])
                    except:
                        try:
                            gu(sha[-1])
                        except Exception as e:
                            print(location, "; url didn't go through")
                            print(e)
                            return("")
        org_name='SCI'
        org_web='https://www.soci.org/'
        org_pro=''   
        if ticketlist=='':
            ticket_list=ticketlist
        else:
            ticket_list=json.dumps([ticketlist],ensure_ascii=False)
        event_info=even
        time_=''
        logo=''
        sponsors=''
        if rack==[] or rack=='':
            agendalist=''
        else:
            agendalist=json.dumps(rack,ensure_ascii=False)
        type_=''
        category=''
        if email=='':
            mail_=''
        else:
            mail_=json.dumps([email],ensure_ascii=False)
        if speak=='':
            Speakerlist=''
        else:
            Speakerlist=json.dumps([speak],ensure_ascii=False)
        event_web=links
        #*********
        try:
            ven=' '.join(soup2.find('p',{'class':'location'}).text.split())
        except:
            ven=''
        if 'Online' in ven or 'Web' in web or ven=='Web':
            ven=''
            on_off='1'
        #*******
        if ven=='':
            venue=''
            city=''
            country=''
            googlePlaceUrl=''
        else:
            venue=ven
            
            try:
                city=ven.split(',')[-2].strip()
            except:
                try:
                    city=ven.split(',')[-1].strip()
                except:
                    city=address
                    
            country=countr(city)
            googlePlaceUrl=get_google_map_url(venue)
        sma.append([links,title,start_date,end_date,time_,event_info,ticket_list,
                     org_pro,org_name,org_web,logo,sponsors,agendalist,
                     type_,category,city,country,venue,event_web,googlePlaceUrl,mail_,
                     Speakerlist,on_off])

print('Done')
```
