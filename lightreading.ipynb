```python
light = []
for i in cards:
    #date = i.find('span',{'class':'date'}).text
    try:
        tutu = i.find('a').text.strip()
    except:
        tutu=''
    
    link=i.find('a')['href']
    #data.append([date,title,link])
    if link=='':
        title=tutu
        dat=''
        info=''
        speakers=''
        sponsor=''
        org_na=''
        org_w=''
        locat=''
        event_w=''
        tim=''
    else:
        driver.get(link)
        soupn = bs(driver.page_source,'lxml')

        #4
        crayze=driver.find_element_by_xpath('/html/body').get_attribute('outerHTML').split('>')[0]
        if 'class' in crayze:
            metho=crayze.split('class')[1].split('"')[1]
        else:
            metho=driver.find_element_by_xpath('/html/body').get_attribute('outerHTML').replace('<','>').split('>')[1]



        if metho=='body':
            over=soupn.find('div',{'id':'tagline-wrap'})

            try:
                title=over.find('span',{'name':'title'}).text
            except:
                title=''
            try:
                dat=over.find('span',{'name':'date'}).text
            except:
                dat=''
            try:
                info=over.find('div',{'class':'summary-section section'}).find('span',{'class':'summary'}).find_all('p')[0].text
            except:
                info=''

            speakers=''
            sponsor=''
            org_na=''
            org_w=''
            locat=''
            event_w=''
            tim=''
            #title, date, time, event_info, speakers,sponsors,org_nam,org_web,locat

              #title,date,org_nam,speakers,sponsors,time,event_info,org_web,locat


        elif 'home blog' in metho:# or 'page' in metho:
            soupn = bs(driver.page_source,'lxml')
            omerg=soupn.find('section',{'id':'home-banner-section'})
            ban=soupn.find('section',{'id':'intro-section'})
            spik=soupn.find('section',{'id':'speaker-section'})
            log=soupn.find('div',{'class':'logo-background'})
            spurn=soupn.find('section',{'id':'sponsor-section'}).find('div',{'class':'home-sponsor-group'}).find_all('a')#['alt']
            try:
                title=omerg.find('div',{'class':'tagline-wrap'}).find('h2').text.replace('  ','').replace('\n',', ')
            except:
                title=''

            try:
                dat=omerg.find('div',{'class':'tagline-wrap'}).find('h3').text.replace('  ','').replace('\n',', ')
            except:
                dat=''

            ########
            try:
                info=ban.find('p').text.replace('\n\n',' ')
            except:
                info=''
            #!!!!!!!!!!!!!!!!!!!!!!!!!!############################

            try:
                sp=spik.find('div',{'class':'owl-carousel owl-theme'}).find('div',{'class':'owl-wrapper'}).find_all('p')
                largo=spik.find('div',{'class':'owl-carousel owl-theme'}).find('div',{'class':'owl-wrapper'}).find_all('a')#['href']
                ##############################
                shi=[[da['href']] for da in largo]
                pid=[(q.text.split('\n')[1]+';'+q.text.split('\n')[2]+', '+q.text.split('\n')[3]).split(';') for q in sp]
                speakers=[pid[y]+shi[y] for y in range(0,len(pid))]
            except:
                speakers=''

            try:
                sponsor=[j['title'] for j in spurn]
            except:
                sponsor=''
            try:
                org_na=log.find('img')['alt']
            except:
                org_na=''
            tim=''
            org_w=''
            locat=''
            event_w=''
            #sponsor=spon

            #title,date,time,event_info,speakers,sponsors,org_nam,org_w,locat




            #org=soupn.find('h2').text#['alt']

        if 'page' in metho:
            soupn = bs(driver.page_source,'lxml')
            twa=soupn.find('section',{'class':'section section--hero'})
            org=soupn.find('header',{'id':'pageheader'})   
            sp=soupn.find('div',{'class':'statement__content'})
            spo=soupn.find('section',{'class':'section section--bg bg-grey-lightest'})

            try:
                title=twa.find('h2').text
            except:
                title=''
            try:
                dat=twa.find('h3').text.replace('\n','').strip()
            except:
                dat=''

            try:
                org_na=org.find('img')['alt'] 
            except:
                org_na=''

            try:
                speakers=sp.find('div',{'class':'statement__content__footer'}).text.replace('  ','').split('\n')[:2]
            except:
                speakers=''

            try:
                spon=spo.find('div',{'class':'grid grid--center grid--huge'}).find_all('a')#,{'title':'grid__cell 4/12 4/12--sm 2/12--md'})
                sponsor=[a['href'].split('www')[1].split('.')[1] for a in spon if a['href'].startswith('http')]
            except:
                sponsor=''

            tim=''
            info=''
            locat=''
            org_w=''
            event_w=link
         #title,date,org_nam,speakers,sponsors,time,event_info,org_web,locat,event_w
            ##title,dat,org_nam,event_info,event_web,speakers,locat,time,org_web,sponsors

        if metho=='windows chrome loaded':
            soupn = bs(driver.page_source,'lxml')
            dtlo=soupn.find('div',{'class':'container-fluid'})
            orag=soupn.find('div',{'class':'DivisionalFooter__Division-sc-v1hex4-0 hYTYwy'})
            sp=soupn.find('section',{'class':'SiteBuilderModule__Section-sc-1ari584-0 promoted-speakers-section'})

            titinev=soupn.find('section',{'class':'home-video-section'})
            if str(titinev)=='None':
                titinev=soupn.find('section',{'class':re.compile('SiteBuilderModule__Section.+')})

            ######################################################################################


            try:
                org_na=orag.find('img')['alt'][:-5]
            except:
                org_na=''
            try:
                org_w=orag.find('a')['href']
            except:
                org_w=''

            try:
                dat=dtlo.find('div',{'class':re.compile('EventVenue__Date-.+')}).text
            except:
                dat=''

            try:
                locat=dtlo.find('span',{'class':re.compile('EventVenue__Address-.+')}).text
            except AttributeError:
                try:
                    locat=dtlo.find('span',{'class':'right-col'}).text
                except:
                    locat=''

            try:
                title=titinev.find('h1').text
            except:
                title=titinev.find('h2').text

            try:
                info=titinev.find('h3',{'class':re.compile('HeroModuleSubTitle.+')}).text
            except AttributeError:
                try:
                    info=titinev.find('div',{'class':re.compile('Paragraph__Paragraph.+')}).text
                except:
                    info=''

            try:
                event_w=titinev.find('a',{'class':'c-cta c-cta--accent c-cta--large c-cta--white'})['href']
            except:
                event_w=link
        ################

            speakers=[]
            # if str(sp)=='none':
            #         speakers=''
            # else:
            try:
                spik=sp.find_all('div',{'class':'col-xs-12 col-sm-3'})

                for jq in spik:
                    try:
                        dam=jq.find('h3',{'data-testid':'speaker-name'}).text
                    except:
                        dam=''
                    try:
                        cam=jq.find('h4',{'data-testid':'speaker-job'}).text    
                    except:
                        cam=''
                    try:
                        bam=jq.find('h4',{'data-testid':'speaker-company'}).text
                    except:
                        ''
                    try:
                        lim=jq.find('a')['href']
                        limk='https://tmt.knect365.com'+lim
                    except:
                        limk=''

                    speakers.append([dam,(cam+', '+bam),limk])
            except:
                speakers=''

            sponsor=''
            tim=''
                ##title,dat,org_nam,event_info,event_web,speakers,locat,time,org_web,sponsors
    #########################DATE###########################
    if dat=='':
        start_date=end_date=''
    else:
        ma=re.search('\d{1,2}.*\W.*\W+(\d{4})',dat)#, maxsplit=0)
        sam=ma.group()
        sama=sam.replace('–','-')
        if '-' in sama:

            st=sama.split('-')[0].strip().replace(',','')
            en=sama.split('-')[1].strip().replace(',','')

            if any(c.isalpha() for c in st)==False:
                pa=re.search('[A-Sa-z]+\W+(\d{4})',en)#, maxsplit=0)
                sapa=pa.group()
                start=st+' '+sapa
                end=en

            elif re.search('\d{1,2}\s+[A-Sa-z]{3,9}',st):
                pa=re.search('(\d{4})',en)#, maxsplit=0)
                sapa=pa.group()
                start=st+' '+sapa
                end=en


            elif re.search('\d{1,2}\w{2}\s+[A-Sa-z]{3,9}',st):
                print('of')
                figo=re.search('\d{1,2}\w{2}',st)
                riigo=re.search('[A-Sa-z]{3,9}',st)
                amigo=figo.group()
                sarigo=riigo.group()
                sta=re.sub('[a-z]{2}','',amigo)
                pa=re.search('(\d{4})',en)#, maxsplit=0)
                sapa=pa.group()
                start=sta+' '+sarigo+' '+sapa
                pb=re.search('[A-Sa-z]+\s+(\d{4})',en)#, maxsplit=0)
                sapb=pb.group()
                sigo=re.search('\d{1,2}\w{2}',en).group()
                stb=re.sub('[a-z]{2}','',sigo)
                end=stb+' '+sapb

            elif re.search('\d{1,2}\w{2}',st):

                sta=re.sub('[a-z]{2}','',st)
                pa=re.search('[A-Sa-z]+\s+(\d{4})',en)#, maxsplit=0)
                sapa=pa.group()
                start=sta+' '+sapa
                ema=re.search('\d{1,2}',en)
                ama=ema.group()
                end=re.sub(re.search('(\d{1,2}\w{2})',en).group(),ama,en)    

            else:
                start=end=''
            if start=='':
                start_date=end_date=''
            else:
                pick=[start, end]
                spl_dt_obj = [datetime.strptime(v, '%d %B %Y') for v in pick]
                date_= [z.strftime('%Y-%m-%d') for z in spl_dt_obj]
                start_date=date_[0]
                end_date=date_[1]
        else:
            start_date=end_date=''
    ###################venue,city,country,gpurl#############
    if locat=='':
        if dat!='':
            pat=re.search('(?<=(\d{4})).*',dat).group().strip()
            if len(pat)>=4:
                locat=pat
    ###############COUNTYR,GPU#################
    def country(locale):
        try:
            google_url_for_location="https://www.google.com/search?q="+locale+"&oq="+locale+"&num=1"
            time.sleep(randint(0,3))
            driver.get(google_url_for_location)
            time.sleep(4)
            google_map_url=driver.find_element_by_id('lu_map').click()
            time.sleep(1)
            country=driver.find_element_by_class_name('x3AX1-LfntMc-header-title-VdSJob').text
            return(country)
        except Exception as e:
            print(e)
            return("")
    
    ####################################
    def get_google_map_url(location):
        try:
            google_url_for_location="https://www.google.com/search?q="+location+"&oq="+location+"&num=1"
            time.sleep(randint(0,3))
            driver.get(google_url_for_location)
            time.sleep(4)
            google_map_url=driver.find_element_by_id('lu_map').click()
            time.sleep(1)
            google_map_url=driver.current_url
            return(google_map_url)
        except Exception as e:
            print(e)
            return("")
    
    
    ###############DEFINED#################
    if locat=='':
        city=''
        country=''
        venue=''
        googlePlaceUrl=''
        on_off='1'
    else:
        city=locat
        on_off='0'
        venue=locat
        country=country(city)
        googlePlaceUrl=get_google_map_url(venue)
    ##########################################################
    time_=tim
    event_info=info
    ticket_list=''
    org_pro=''
    org_name=org_na
    org_web=org_w
    event_web=event_w
    category=''
    logo=''
    if sponsor=='':
        sponsors=''
    else:
        sponsors=json.dumps(sponsor)
    agendalist=''
    type_=''
    if speakers=='':
        Speakerlist=''
    else:
        Speakerlist=json.dumps(speakers)
    mail_=''
    




    light.append([link,title,start_date,end_date,time_,event_info,ticket_list,
                 org_pro,org_name,org_web,logo,sponsors,agendalist,
                 type_,category,city,country,venue,event_web,googlePlaceUrl,mail_,
                 Speakerlist,on_off])
    
print('Done')
```
