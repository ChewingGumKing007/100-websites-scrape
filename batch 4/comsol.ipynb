```python
comsol=[]
for m in cards:
    tesla=' '.join(m.find('span',{'class':'event-card-typeBadge'}).text.split())
    tit=' '.join(m.find('a').h2.text.split())
    web=m.find('a')['href']
    if web.startswith('http'):
        link=web
        
    else:
        
        link='https://www.comsol.com'+web
    
    tim=' '.join(m.find('div',{'class':'ec-datetime clearfix'}).find('span',{'class':'ec-location'}).text.split())
    dat=' '.join(m.find('div',{'class':'ec-datetime clearfix'}).find('span',{'class':re.compile('ec-date.+')}).text.split())
    even=' '.join(m.find('div',{'class':'ec-desc'}).text.split())
    #*****************
    for m in ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']:
        if m in dat:
            dat=dat.replace(m,'')
    dat=dat.replace('–','-').replace('|','').replace('to','-').replace('TO','-').strip().replace('Oc-ber','October').replace('OC-BER','OCTOBER')
    if '(' in dat:
        dat=dat.split('(')[0].strip()
    if dat=='':
        start_date=end_date=''
    else:
        mad=dat
        #('September 8, 2022 I The Ritz-Carlton, Beijing',
    if re.search('[A-Sa-z]{3,9}\s?\d{1,2}\s+\d{4}\s?[A-Za-z]+\s?[A-Za-z]+\s?[A-Za-z]+',mad.replace(',','')):
        #print('juice')
        tad=re.search('[A-Sa-z]{3,9}\s?\d{1,2}\s+\d{4}',mad.replace(',','')).group()
        mad=tad
        st=mad.strip().replace(',','').replace('nd','').replace('st','').replace('th','').replace('rd','').replace('ugu','ugust')
        en=mad.strip().replace(',','').replace('nd','').replace('st','').replace('th','').replace('rd','').replace('ugu','ugust')

    elif '-' in mad:
        st=mad.split('-')[0].strip().replace(',','').replace('nd','').replace('st','').replace('th','').replace('rd','').replace('ugu','ugust')
        en=mad.split('-')[1].strip().replace(',','').replace('nd','').replace('st','').replace('th','').replace('rd','').replace('ugu','ugust')

    elif 'and'in mad:
        st=mad.split('and')[0].strip().replace(',','').replace('nd','').replace('st','').replace('th','').replace('rd','').replace('ugu','ugust')
        en=mad.split('and')[1].strip().replace(',','').replace('nd','').replace('st','').replace('th','').replace('rd','').replace('ugu','ugust')

    elif '&' in mad:
        st=mad.split('&')[0].strip().replace(',','').replace('nd','').replace('st','').replace('th','').replace('rd','').replace('ugu','ugust')
        en=mad.split('&')[1].strip().replace(',','').replace('nd','').replace('st','').replace('th','').replace('rd','').replace('ugu','ugust')

    else:
        #print(mad)
        st=mad.strip().replace(',','').replace('nd','').replace('st','').replace('th','').replace('rd','').replace('ugu','ugust')
        en=mad.strip().replace(',','').replace('nd','').replace('st','').replace('th','').replace('rd','').replace('ugu','ugust')

            
    if any(c.isalpha() for c in st)==False:
    #      print('leg')
        pa=re.search('[A-Sa-z]+\W+(\d{4})',en)#, maxsplit=0)
        sapa=pa.group()
        start=st+' '+sapa
        end=en
    #                        02        April      2022
    elif re.search('\d{1,2}\s+[A-Sa-z]{3,9}\s\d{4}',st):
     #   print('awujale')
        start=st
        end=en

    #*                        April       02      2022
    elif re.search('[A-Sa-z]{3,9}\s\d{1,2}\s+\d{4}',st):
    #       print('ala')
        pa=re.search('(\d{4})',st).group()
        sa=re.search('([A-Sa-z]{3,9})',st).group()
        ta=re.search('(\d{1,2})',st).group()
        start=ta+' '+sa+' '+pa
        #
        ba=re.search('(\d{4})',en).group()
        ca=re.search('([A-Sa-z]{3,9})',en).group()
        da=re.search('(\d{1,2})',en).group()
        end=da+' '+ca+' '+ba     

    #                      02     2022
    elif re.search('\d{1,2}\s+\d{4}',en):
    #      print('kum')
        pa=re.search('(\d{4})',en).group()#, maxsplit=0)
        sa=re.search('([A-Sa-z]{3,9})',st).group()
        ta=re.search('(\d{1,2})',st).group()

        start=ta+' '+sa+' '+pa
        #
        ba=re.search('(\d{4})',en).group()
        ca=re.search('([A-Sa-z]{3,9})',st).group()
        da=re.search('(\d{1,2})',en).group()
        end=da+' '+ca+' '+ba

    #                      02         April
    elif re.search('\d{1,2}\s+[A-Sa-z]{3,9}',st):
    #        print('is')
        pa=re.search('(\d{4})',en)#, maxsplit=0)
        sapa=pa.group()
        start=st+' '+sapa
        end=en

    #*                      April           02
    elif re.search('[A-Sa-z]{3,9}\s+\d{1,2}',st):
    #      print('bad')
        pa=re.search('(\d{4})',en).group()#, maxsplit=0)
        sa=re.search('([A-Sa-z]{3,9})',st).group()
        ta=re.search('(\d{1,2})',st).group()

        start=ta+' '+sa+' '+pa
        #
        ba=re.search('(\d{4})',en).group()
        ca=re.search('([A-Sa-z]{3,9})',en).group()
        da=re.search('(\d{1,2})',en).group()
        end=da+' '+ca+' '+ba

    else:
    #     print('shik')
        start=end=''
    if start=='':
        start_date=end_date=''
    else:
        pick=[start, end]
        try:
            spl_dt_obj = [datetime.strptime(v.strip(), '%d %B %Y') for v in pick]
        except:
            spl_dt_obj = [datetime.strptime(v.strip(), '%d %b %Y') for v in pick]
        date_= [z.strftime('%Y-%m-%d') for z in spl_dt_obj]
        start_date=date_[0]
        end_date=date_[1]
        
#     #tim
    tes=tim.split()[-1]
    if any(c.isnumeric() for c in tes)==False:
        z=tes
    else:
        z=''
    ase=tim.replace(z,'').replace('–','-').strip().replace('.','').upper()
    if '-' in ase:
        stt=ase.split('-')[0]
        ett=ase.split('-')[1]
    else:
        stt=ase
        ett=''
    time_st={   
            'type':'general',
            'Start_time':stt,
            'end_time':ett,
            'timezone':z,
            'days':'all'
        }
    time_=json.dumps([time_st])
    driver.get(link)
    time.sleep(5)
    soupn=bs(driver.page_source,'lxml')
   

    fir_div=soupn.find('div')['class']
    if fir_div==['container']:
        try:
            title=' '.join(soupn.find('div',{'class':'overview-section section'}).find('span',{'name':'title'}).text.split())
        except:
            try:
                time.sleep(5)
                soupn=bs(driver.page_source,'lxml')
                title=' '.join(soupn.find('div',{'class':'overview-section section'}).find('span',{'name':'title'}).text.split())
            except:
                time.sleep(5)
                soupn=bs(driver.page_source,'lxml')
                title=' '.join(soupn.find('div',{'class':'overview-section section'}).find('span',{'name':'title'}).text.split())
        dat=' '.join(soupn.find('div',{'class':'overview-section section'}).find('span',{'name':'date'}).text.split())
        tim=' '.join(soupn.find('div',{'class':'overview-section section'}).find('span',{'name':'time'}).text.split())
        info=' '.join(soupn.find('span',{'class':'summary'}).find_all('p')[1].text.split())
        sp=soupn.find('div',{'class':'speaker-fields-container'}).find_all('div',{'class':'section'})
        dupe=[]
        for u in sp:
            na=' '.join(u.find('span',{'class':'sp-name'}).text.split())
            til=' '.join(u.find('span',{'class':'sp-company'}).text.split())+', '+' '.join(u.find('span',{'class':'sp-company'}).text.split())
            lm=''
            dupe.append([na,til,lm])
    elif fir_div==['alert', 'alert-error']:
        title=' '.join(soupn.find('div',{'class':'heading-card-title'}).text.split())
        dat=' '.join(soupn.find('div',{'class':'heading-card-subtitle-text'}).find_all('span')[0].text.split())
        tim=' '.join(soupn.find('div',{'class':'heading-card-subtitle-text'}).find_all('span')[-1].text.split())
        info=' '.join(soupn.find('div',{'class':'ems-event-content content-box'}).text.split())
        try:
            na=' '.join(soupn.find('div',{'class':'evi-speak'}).find('span',{'class':'evi-name'}).text.split())
            t=' '.join(soupn.find('div',{'class':'evi-speak'}).find('span',{'class':'evi-co'}).text.split())
            ln=''
            dupe=[[na,t,ln]]
        except:
            dupe=''
    elif fir_div==['aspNetHidden']:
        title=' '.join(soupn.find('div',{'class':'BOX_Content'}).h1.text.split())
        dash=' '.join(soupn.find('div',{'class':'BOX_Content'}).b.text.split()).replace('Mon','').replace('Tue','').replace('Wed','').replace('Thur','').replace('Fri','').replace('Sat','').replace('Sun','').replace(',','')
        dat=re.search('[A-Sa-z]{3,9}\s\d{1,2}\s+\d{4}',dash).group()
        tim=dash.replace(dat,'').strip()
        info=' '.join(soupn.find('div',{'class':'BOX_Content'}).find('p').text.split())
        dupe=''
    if dupe=='' or dupe==[]:
        Speakerlist=''
    else:
        tweak=[]
        for em in dupe:
            speak={
                'name':em[0],
                'title':em[1],
                'link':em[2]
            }
            tweak.append(speak)
        Speakerlist=json.dumps(tweak,ensure_ascii=False)
        
        
    def gu(luc):
        google_url_for_location="https://www.google.com/search?q="+luc+"&oq="+luc+"&num=1"
        time.sleep(randint(0,3))
        driver.get(google_url_for_location)
        time.sleep(4)
        try:
            google_map_url=driver.find_element_by_id('lu_map').click()
        except:
            try:
                google_map_url=driver.find_element_by_class_name('Xm7sWb').click()
            except:
                google_map_url=driver.find_element_by_class_name('Lx2b0d').click()
        time.sleep(1)
        google_map_url=driver.current_url
#                 print(google_map_url)
        return(google_map_url)
        ######################################
    def get_google_map_url(location):
        try:
            return(gu(location))
        except:
            sha=location.split(',')
            try:
                gu(sha[-3])
            except:
                try:
                    gu(sha[-2])
                except:
                    try:
                        gu(sha[-1])
                    except Exception as e:
                        print(location, "; url didn't go through")
                        print(e)
                        return("")


    def countr(locale):
        try:
            google_url_for_location="https://www.google.com/search?q="+locale+"&oq="+locale+"&num=1"
            time.sleep(randint(0,3))
            driver.get(google_url_for_location)
            time.sleep(4)
            try:
                google_map_url=driver.find_element_by_id('lu_map').click()
            except:
                try:
                    google_map_url=driver.find_element_by_class_name('Xm7sWb').click()
                except:
                    google_map_url=driver.find_element_by_class_name('Lx2b0d').click()
            time.sleep(1)
            try:
                country=driver.find_element_by_class_name('x3AX1-LfntMc-header-title-VdSJob').text
            except:
                country=driver.find_element_by_class_name('bwoZTb').text
#                 print(country)
            return(country)
        except Exception as e:
            print(locale, ";country didn't go through")
            print(e)
            return("")

    ticket_list=''
    org_name='COMSOL'
    org_web='https://www.comsol.jp/'
    org_pro=''
    logo=''    
    sponsor=''
    agendalist=''
    type_=''
    category=''
    event_info=info
#     mass='ratings.request@spglobal.com'
    mail_=''
    event_web=link
    if tesla=='Webinar':
        on_off='1'
        city=''
        venue=''
        country=''
        googlePlaceUrl=''


    comsol.append([link,title,start_date,end_date,time_,event_info,ticket_list,
                     org_pro,org_name,org_web,logo,sponsor,agendalist,
                     type_,category,city,country,venue,event_web,googlePlaceUrl,mail_,
                     Speakerlist,on_off])

print('Done')
```
