```python
time.sleep(5)
soup=bs(driver.page_source,'lxml')
cards=soup.find_all('div',{'class':'event-block'})
print('The total number of events is',len(cards))
```

```python
amco=[]

sag=1
tram=1
for i in cards:
    ritle=' '.join(i.find('h4',{'class':'event-title'}).text.split())
    link=i.find('div',{'class':'event-block-body'}).find('a')['href']

    #pier=i.find('div',{'class':'block-meta event-date-location'})
    da=' '.join(i.find('p',{'class':'event-date'}).text.split())
    try:
        lo=' '.join(i.find('p',{'class':'event-location'}).text.split())
    except:
        
        lo='virtual'
    #info=' '.join(i.find('div',{'class':'mt-3 mb-3 h6'}).text.split())
    driver.get(link)
    soupn=bs(driver.page_source,'lxml')
    
    try: 
        ase=' '.join(soupn.find('h5',{'class':'event-title-prefix'}).text.split())
    except:
        ase=''
    asa=' '.join(soupn.find('h1',{'class':'event-title'}).text.split())
    title=ase+asa
    #try:
    dat_dur=' '.join(soupn.find('span',{'class':'event-date'}).text.split())
    
    dat=dat_dur.replace('–','-')
    if '(' in dat:
        dat=dat.split('(')[0].strip()
    if dat=='':
        start_date=end_date=''
    else:
        mad=dat
        #try:
#             ma=re.search('([A-Sa-z]+\W+\d{1,2}.+\W+\d{4}|\d{1,2}.+\W+\d{4})',_date)#, maxsplit=0)
#             mad=ma.group()

        if '-' in mad:
            st=mad.split('-')[0].strip().replace(',','').replace('nd','').replace('st','').replace('th','').replace('rd','').replace('ugu','ugust')
            en=mad.split('-')[1].strip().replace(',','').replace('nd','').replace('st','').replace('th','').replace('rd','').replace('ugu','ugust')

        elif 'and'in mad:
            st=mad.split('and')[0].strip().replace(',','').replace('nd','').replace('st','').replace('th','').replace('rd','').replace('ugu','ugust')
            en=mad.split('and')[1].strip().replace(',','').replace('nd','').replace('st','').replace('th','').replace('rd','').replace('ugu','ugust')

        elif '&' in mad:
            st=mad.split('&')[0].strip().replace(',','').replace('nd','').replace('st','').replace('th','').replace('rd','').replace('ugu','ugust')
            en=mad.split('&')[1].strip().replace(',','').replace('nd','').replace('st','').replace('th','').replace('rd','').replace('ugu','ugust')

        else:
            st=mad.strip().replace(',','').replace('nd','').replace('st','').replace('th','').replace('rd','').replace('ugu','ugust')
            en=mad.strip().replace(',','').replace('nd','').replace('st','').replace('th','').replace('rd','').replace('ugu','ugust')
        #print(date__,st,en)
        ####################
        if any(c.isalpha() for c in st)==False:
        #      print('leg')
            pa=re.search('[A-Sa-z]+\W+(\d{4})',en)#, maxsplit=0)
            sapa=pa.group()
            start=st+' '+sapa
            end=en
        #                        02        April      2022
        elif re.search('\d{1,2}\s+[A-Sa-z]{3,9}\s\d{4}',st):
         #   print('awujale')
            start=st
            end=en

        #*                        April       02      2022
        elif re.search('[A-Sa-z]{3,9}\s\d{1,2}\s+\d{4}',st):
        #       print('ala')
            pa=re.search('(\d{4})',st).group()
            sa=re.search('([A-Sa-z]{3,9})',st).group()
            ta=re.search('(\d{1,2})',st).group()
            start=ta+' '+sa+' '+pa
            #
            ba=re.search('(\d{4})',en).group()
            ca=re.search('([A-Sa-z]{3,9})',en).group()
            da=re.search('(\d{1,2})',en).group()
            end=da+' '+ca+' '+ba     

        #                      02     2022
        elif re.search('\d{1,2}\s+\d{4}',en):
        #      print('kum')
            pa=re.search('(\d{4})',en).group()#, maxsplit=0)
            sa=re.search('([A-Sa-z]{3,9})',st).group()
            ta=re.search('(\d{1,2})',st).group()

            start=ta+' '+sa+' '+pa
            #
            ba=re.search('(\d{4})',en).group()
            ca=re.search('([A-Sa-z]{3,9})',st).group()
            da=re.search('(\d{1,2})',en).group()
            end=da+' '+ca+' '+ba

        #                      02         April
        elif re.search('\d{1,2}\s+[A-Sa-z]{3,9}',st):
        #        print('is')
            pa=re.search('(\d{4})',en)#, maxsplit=0)
            sapa=pa.group()
            start=st+' '+sapa
            end=en

        #*                      April           02
        elif re.search('[A-Sa-z]{3,9}\s+\d{1,2}',st):
        #      print('bad')
            pa=re.search('(\d{4})',en).group()#, maxsplit=0)
            sa=re.search('([A-Sa-z]{3,9})',st).group()
            ta=re.search('(\d{1,2})',st).group()

            start=ta+' '+sa+' '+pa
            #
            ba=re.search('(\d{4})',en).group()
            ca=re.search('([A-Sa-z]{3,9})',en).group()
            da=re.search('(\d{1,2})',en).group()
            end=da+' '+ca+' '+ba

        else:
        #     print('shik')
            start=end=''
        if start=='':
            start_date=end_date=''
        else:
            pick=[start, end]
            spl_dt_obj = [datetime.strptime(v, '%d %B %Y') for v in pick]
            date_= [z.strftime('%Y-%m-%d') for z in spl_dt_obj]
            start_date=date_[0]
            end_date=date_[1]
    
    
    try:
        inr=' '.join(soupn.find('div',{'class':'col-xs-12 col-md-8 col-md-offset-2'}).text.split())
        if 'ABOUT THIS CONFERENCE' in inr:
            event_info=inr.replace('ABOUT THIS CONFERENCE','').strip()
        elif 'ABOUT THIS CONFERENCE'.title() in inr:
            event_info=inr.replace('ABOUT THIS CONFERENCE'.title(),'').strip()
        elif 'ABOUT THIS CONFERENCE'.lower() in inr:
            event_info=inr.replace('ABOUT THIS CONFERENCE'.lower,'').strip()
        else:
            event_info=' '.join(soupn.find('div',{'class':'row vertical-spacer-half'}).text.split())
    except:
        try:
            event_info=' '.join(soupn.find('div',{'class':'row vertical-spacer-half'}).text.split())
        except:
            event_info=' '.join(soupn.find('p',{'class':'header-copy'}).text.split())
    try:
        sps=soupn.find('div',{'class':'content-block-row row sponsor-group-block-row'}).find_all('li',{'class':'partner-block'})
        spon=[a.find('a')['href'] for a in sps]
        spons=[b.split('/')[-2] for b in spon]
        sponsor=json.dumps(spons)
    except:
        sponsor=''
    def gu(luc):
        google_url_for_location="https://www.google.com/search?q="+luc+"&oq="+luc+"&num=1"
        time.sleep(randint(0,3))
        driver.get(google_url_for_location)
        time.sleep(4)
        try:
            google_map_url=driver.find_element_by_id('lu_map').click()
        except:
            try:
                google_map_url=driver.find_element_by_class_name('Xm7sWb').click()
            except:
                google_map_url=driver.find_element_by_class_name('Lx2b0d').click()
        time.sleep(1)
        google_map_url=driver.current_url
#                 print(google_map_url)
        return(google_map_url)
        ######################################
    def get_google_map_url(location):
        try:
            return(gu(location))
        except:
            sha=location.split(',')
            try:
                gu(sha[-3])
            except:
                try:
                    gu(sha[-2])
                except:
                    try:
                        gu(sha[-1])
                    except Exception as e:
                        print(location, "; url didn't go through")
                        print(e)
                        return("")


    def countr(locale):
        try:
            google_url_for_location="https://www.google.com/search?q="+locale+"&oq="+locale+"&num=1"
            time.sleep(randint(0,3))
            driver.get(google_url_for_location)
            time.sleep(4)
            try:
                google_map_url=driver.find_element_by_id('lu_map').click()
            except:
                try:
                    google_map_url=driver.find_element_by_class_name('Xm7sWb').click()
                except:
                    google_map_url=driver.find_element_by_class_name('Lx2b0d').click()
            time.sleep(1)
            try:
                country=driver.find_element_by_class_name('x3AX1-LfntMc-header-title-VdSJob').text
            except:
                country=driver.find_element_by_class_name('bwoZTb').text
#                 print(country)
            return(country)
        except Exception as e:
            print(locale, ";country didn't go through")
            print(e)
            return("")    
     #**********
    
    
        
    check=' '.join(soupn.find('a',{'class':'event-delivery-method'}).text.split())
    
    try:
        loc=' '.join(soupn.find('span',{'class':'event-location'}).text.split())
    except:
        loc=''
    
    if 'Virtual' in check or loc=='' :
        tram+=1
        on_off='1'
        city=''
        country=''
        venue=''
        googlePlaceUrl=''
    else:
        on_off='0'
        try:
            city=loc.split(',')[-2]
            xtra=city+', '+loc.split(',')[-1]
        except:
            city=loc.split(',')[0]
        country=countr(city)
        if country=='':
            country=countr(xtra)
        venue=loc
        googlePlaceUrl=get_google_map_url(venue)
        if googlePlaceUrl=='':
            googlePlaceUrl=get_google_map_url(xtra)
    ticket_list=''
    org_name='American Conference Institute'
    org_web='https://www.americanconference.com/'
    org_pro='Our conferences connect the power of people with the power of information, a powerful combination for business growth and success. Each conference, summit or executive briefing is tailored to the sector, topic and audience need.'
    logo=''    
    #sponsor=''
    agendalist=''
    type_=''
    category=''
    Speakerlist='' 
    time_=''
    event_web=link
    mail_=json.dumps(['customerservice@americanconference.com'])
    amco.append([link,title,start_date,end_date,time_,event_info,ticket_list,
                 org_pro,org_name,org_web,logo,sponsor,agendalist,
                 type_,category,city,country,venue,event_web,googlePlaceUrl,mail_,
                 Speakerlist,on_off])
    sag+=1
    for d in range(0,50*100,100):
        if sag==d:
            print(d,'events scraped')

print('Done')
print('There are ',tram,' number of online events.')
```
