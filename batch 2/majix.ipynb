```python
majix=[]
soup = bs(driver.page_source,'lxml')
cards=soup.find_all('div',{'class':'All_courses_card__Lwu1Z card'})
for i in cards:
    link='https://mindmajix.com'+i.find('a')['href']
    driver.get(link)
    time.sleep(3)
    soupn = bs(driver.page_source,'lxml')
    #**************8888
    on=soupn.find('div',{'class':re.compile('LiveVirtualTraining_datebox.+')}).find_all('div',{'class':'row'})
    sue=[o.get_text(separator=' ') for o in on][1:]#.remove(' ')
    yr=date.today().year
    for z in sue:
        dat=re.search('[A-Za-z]{3}\W+\d{2}\W+[A-Za-z]{3}\W+\d{2}',z).group()
        sat=dat.split('-')
        sa=[(u+' '+str(yr)).replace('  ',' ').replace('  ',' ').strip() for u in sat]
        spl_dt_obj = [datetime.strptime(v, '%b %d %Y') for v in sa]
        dates_= [z.strftime('%Y-%m-%d') for z in spl_dt_obj]
        start_date=dates_[0]
        end_date=dates_[1]
        tim=re.search('\d{2}:\d{2}\W+[A-Z]{2}\W+[A-Z]{3}',z).group().replace(' AM','AM').replace(' PM','PM').split()+['']
        if tim=='':
            time_=''
        else:
            time_st=[{'type': 'general',
                     'start_time':tim[0],
                     'end_time': tim[2],
                     'timezone': tim[1],
                     'days': 'all'}]
            time_=json.dumps(time_st)
        
        #*****************
        title=soupn.find('div',{'class':'Banner_hero_wrap__qLL4V'}).h1.text
        try:
            event_info=top=soupn.find('div',{'class':re.compile('Curriculum_description.+')}).text
        except:
            event_info=''
        #**************************
        try:
            skerlist=soupn.find('section',{'class':re.compile('instructor Instructor_instructor.+')}).find('table').tbody.find_all('tr')[:3]
        except:
            skerlist=''
        if skerlist=='':
            Speakerlist=''
        else:
            ase=[i.td.text for i in skerlist]
            speakers=[{'name': ase[0],
                     'title':'years experience with; '.join(ase[1:]),
                     'link': ''}]
            Speakerlist=json.dumps(speakers)
        #****************
        try:
            swim=soupn.find('div',{'class':re.compile('LiveVirtualTraining_detail.+')}).find('p',{'class':re.compile('LiveVirtualTraining_price.+')}).text.replace(',','')
        except:
            swim=''
        if '$' in swim:
            sia=swim.replace('$','').strip()
            if int(sia)>0:
                ty='paid'
                am=sia
                cu='$'
            else:
                ty='free'
                am=''
                cu=''
        elif '£' in swim:
            sia=swim.replace('£','').strip() 
            if int(sia)>0:
                ty='paid'
                am=sia
                cu='£'
            else:
                ty='free'
                am=''
                cu=''

        elif '€' in swim:
            sia=swim.replace('€','').strip()
            if int(sia)>0:
                ty='paid'
                am=sia
                cu='€'
            else:
                ty='free'
                am=''
                cu=''

        else:
            ty=''
            am=''
            cu=''
        if ty=='':
            ticket_list=''
        else:
            ticketlist=[{'type': ty,
                         'price':am,
                         'currency': cu}]
            ticket_list=json.dumps(ticketlist)
        # if ticketlist=='':
        #     ticket_list=ticketlist
        # else:

        #ticket_list

         ###############COUNTYR,GPU#################
        def country(locale):
            try:
                google_url_for_location="https://www.google.com/search?q="+locale+"&oq="+locale+"&num=1"
                time.sleep(randint(0,3))
                driver.get(google_url_for_location)
                time.sleep(4)
                google_map_url=driver.find_element_by_id('lu_map').click()
                time.sleep(1)
                country=driver.find_element_by_class_name('x3AX1-LfntMc-header-title-VdSJob').text
                return(country)
            except Exception as e:
                print(e)
                return("")

        ####################################
        def get_google_map_url(location):
            try:
                google_url_for_location="https://www.google.com/search?q="+location+"&oq="+location+"&num=1"
                time.sleep(randint(0,3))
                driver.get(google_url_for_location)
                time.sleep(4)
                google_map_url=driver.find_element_by_id('lu_map').click()
                time.sleep(1)
                google_map_url=driver.current_url
                return(google_map_url)
            except Exception as e:
                print(e)
                return("")

        ven=''
        on=soupn.find('div',{'class':re.compile('LiveVirtualTraining_training.+')})
        if 'Online' in str(on):
            on_off='1'
            city=''
            country=''
            venue=''
            googlePlaceUrl=''
        else:
            on_off='0'
            city=ven
            venue=ven
            country=country(city)
            googlePlaceUrl=get_google_map_url(venue)

    
        org_pro='MindMajix is a live and interactive e-learning platform that offers professional online education to help companies improve their profitability drastically. We provide several cyber training in various fields which includes but not limited to: DevOps, Business Analytics, Database administration, VMware, Selenium, Amazon Web Services, Bigdata, Python, Identity Management, IT service, Management, NoSQL DB Courses, IT Networking and Server Administration Courses.'
        org_web='https://mindmajix.com/'
        org_name='MindMajix Technologies'
        mass=['info@mindmajix.com']
        mail_=json.dumps(mass)
        logo=''
        sponsors=''
        agendalist=''
        type_=''
        category=''
        Speakerlist=''
        event_web=link
    majix.append([link,title,start_date,end_date,time_,event_info,ticket_list,
                 org_pro,org_name,org_web,logo,sponsors,agendalist,
                 type_,category,city,country,venue,event_web,googlePlaceUrl,mail_,
                 Speakerlist,on_off])
        
print('Done')
```
