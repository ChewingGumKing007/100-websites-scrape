```python
opal=[]
for i in cards:
    ritle=' '.join(i.find('a',{'class':'conf_title'}).text.split())
    link=i.find('a')['href']
    dat=' '.join(i.find('span',{'class':'date'}).text.split())
 
    lo=' '.join(i.find('span',{'class':'adress'}).text.split())
#     info=' '.join(i.find('div',{'class':'mt-3 mb-3 h6'}).text.split())
    driver.get(link)
    soupn=bs(driver.page_source,'lxml')
#     try:
    dat_dur=' '.join(soupn.find('p').get_text(separator='{').split()).split('{')[-1].strip()
    #print(dat_dur)
    da=dat_dur.replace('–','-')
    if '(' in da:
        da=dat.split('(')[0].strip()
    if da=='':
        start_date=end_date=''
    else:
        mad=da
        #try:
#             ma=re.search('([A-Sa-z]+\W+\d{1,2}.+\W+\d{4}|\d{1,2}.+\W+\d{4})',_date)#, maxsplit=0)
#             mad=ma.group()

        if '-' in mad:
            st=mad.split('-')[0].strip().replace(',','').replace('nd','').replace('st','').replace('th','').replace('rd','').replace('ugu','ugust')
            en=mad.split('-')[1].strip().replace(',','').replace('nd','').replace('st','').replace('th','').replace('rd','').replace('ugu','ugust')

        elif 'and'in mad:
            st=mad.split('and')[0].strip().replace(',','').replace('nd','').replace('st','').replace('th','').replace('rd','').replace('ugu','ugust')
            en=mad.split('and')[1].strip().replace(',','').replace('nd','').replace('st','').replace('th','').replace('rd','').replace('ugu','ugust')

        elif '&' in mad:
            st=mad.split('&')[0].strip().replace(',','').replace('nd','').replace('st','').replace('th','').replace('rd','').replace('ugu','ugust')
            en=mad.split('&')[1].strip().replace(',','').replace('nd','').replace('st','').replace('th','').replace('rd','').replace('ugu','ugust')

        else:
            st=mad.strip().replace(',','').replace('nd','').replace('st','').replace('th','').replace('rd','').replace('ugu','ugust')
            en=mad.strip().replace(',','').replace('nd','').replace('st','').replace('th','').replace('rd','').replace('ugu','ugust')
        #print(date__,st,en)
        ####################
        if any(c.isalpha() for c in st)==False:
        #      print('leg')
            pa=re.search('[A-Sa-z]+\W+(\d{4})',en)#, maxsplit=0)
            sapa=pa.group()
            start=st+' '+sapa
            end=en
        #                        02        April      2022
        elif re.search('\d{1,2}\s+[A-Sa-z]{3,9}\s\d{4}',st):
         #   print('awujale')
            start=st
            end=en

        #*                        April       02      2022
        elif re.search('[A-Sa-z]{3,9}\s\d{1,2}\s+\d{4}',st):
        #       print('ala')
            pa=re.search('(\d{4})',st).group()
            sa=re.search('([A-Sa-z]{3,9})',st).group()
            ta=re.search('(\d{1,2})',st).group()
            start=ta+' '+sa+' '+pa
            #
            ba=re.search('(\d{4})',en).group()
            ca=re.search('([A-Sa-z]{3,9})',en).group()
            da=re.search('(\d{1,2})',en).group()
            end=da+' '+ca+' '+ba     

        #                      02     2022
        elif re.search('\d{1,2}\s+\d{4}',en):
        #      print('kum')
            pa=re.search('(\d{4})',en).group()#, maxsplit=0)
            sa=re.search('([A-Sa-z]{3,9})',st).group()
            ta=re.search('(\d{1,2})',st).group()

            start=ta+' '+sa+' '+pa
            #
            ba=re.search('(\d{4})',en).group()
            ca=re.search('([A-Sa-z]{3,9})',st).group()
            da=re.search('(\d{1,2})',en).group()
            end=da+' '+ca+' '+ba

        #                      02         April
        elif re.search('\d{1,2}\s+[A-Sa-z]{3,9}',st):
        #        print('is')
            pa=re.search('(\d{4})',en)#, maxsplit=0)
            sapa=pa.group()
            start=st+' '+sapa
            end=en

        #*                      April           02
        elif re.search('[A-Sa-z]{3,9}\s+\d{1,2}',st):
        #      print('bad')
            pa=re.search('(\d{4})',en).group()#, maxsplit=0)
            sa=re.search('([A-Sa-z]{3,9})',st).group()
            ta=re.search('(\d{1,2})',st).group()

            start=ta+' '+sa+' '+pa
            #
            ba=re.search('(\d{4})',en).group()
            ca=re.search('([A-Sa-z]{3,9})',en).group()
            da=re.search('(\d{1,2})',en).group()
            end=da+' '+ca+' '+ba

        else:
        #     print('shik')
            start=end=''
        if start=='':
            start_date=end_date=''
        else:
            pick=[start, end]
            spl_dt_obj = [datetime.strptime(v, '%d %B %Y') for v in pick]
            date_= [z.strftime('%Y-%m-%d') for z in spl_dt_obj]
            start_date=date_[0]
            end_date=date_[1]
    
    title=' '.join(soupn.find('h1').text.split())

    
    #try:
    event_info=' '.join(soupn.find('div',{'class':'conference-details'}).p.text.split())
    if event_info=='' or event_info==' ':
        event_info=' '.join(soupn.find('div',{'class':'conference-details'}).find_all('p')[1].text.split())
    
    #try:
    sps=soupn.find('div',{'class':'brands'}).find_all('li')
    tic=[a.find('i')['style'].replace('-Logo','').replace('-Web','').replace('-web','').split('/')[-1].split('.')[-2] for a in sps]
    if tic==[] or tic=='' or tic=='-' or tic==' ':
        sponsor=''
    else:
        sponsor=json.dumps(tic)
    #print(sponsor)
#     except:
#         sps=''
    #******
    def gu(luc):
        google_url_for_location="https://www.google.com/search?q="+luc+"&oq="+luc+"&num=1"
        time.sleep(randint(0,3))
        driver.get(google_url_for_location)
        time.sleep(4)
        try:
            google_map_url=driver.find_element_by_id('lu_map').click()
        except:
            try:
                google_map_url=driver.find_element_by_class_name('Xm7sWb').click()
            except:
                google_map_url=driver.find_element_by_class_name('Lx2b0d').click()
        time.sleep(1)
        google_map_url=driver.current_url
#                 print(google_map_url)
        return(google_map_url)
        ######################################
    def get_google_map_url(location):
        try:
            return(gu(location))
        except:
            sha=location.split(',')
            try:
                gu(sha[-3])
            except:
                try:
                    gu(sha[-2])
                except:
                    try:
                        gu(sha[-1])
                    except Exception as e:
                        print(location, "; url didn't go through")
                        print(e)
                        return("")


    def countr(locale):
        try:
            google_url_for_location="https://www.google.com/search?q="+locale+"&oq="+locale+"&num=1"
            time.sleep(randint(0,3))
            driver.get(google_url_for_location)
            time.sleep(4)
            try:
                google_map_url=driver.find_element_by_id('lu_map').click()
            except:
                try:
                    google_map_url=driver.find_element_by_class_name('Xm7sWb').click()
                except:
                    google_map_url=driver.find_element_by_class_name('Lx2b0d').click()
            time.sleep(1)
            try:
                country=driver.find_element_by_class_name('x3AX1-LfntMc-header-title-VdSJob').text
            except:
                country=driver.find_element_by_class_name('bwoZTb').text
#                 print(country)
            return(country)
        except Exception as e:
            print(locale, ";country didn't go through")
            print(e)
            return("")    
     #**********
    
    loc=' '.join(soupn.find('p').get_text(separator='{').split()).split('{')[0].strip()
    if loc!='' or loc!=' ' or 'irtual' not in loc or 'nline' not in loc or 'web' not in loc:
        if ',' in loc:
            city=loc.split(',')[-2].strip()
            xtra=city+', '+loc.split(',')[-1].strip()
        else:
            city=loc
            
        venue=loc
        country=countr(city)
        if country=='':
            country=countr(xtra)
        on_off='0'
        googlePlaceUrl=get_google_map_url(venue)
        if googlePlaceUrl=='':
            googlePlaceUrl=get_google_map_url(xtra)
    else:
        venue=''
        country=''
        on_off='1'
        city=''
        googlePlaceUrl=''
    ticket_list=''
    org_name='Opal Group'
    org_web='https://opalgroup.net/'
    org_pro='Opal Group is a global conference organizer catering to top executives and decision makers throughout various industries.'
    logo=''    
    sponsor=''
    agendalist=''
    type_=''
    category=''
    Speakerlist='' 
    time_=''
    event_web=link
    mail_=json.dumps(['feedback@opalgroup.net'])
    opal.append([link,title,start_date,end_date,time_,event_info,ticket_list,
                 org_pro,org_name,org_web,logo,sponsor,agendalist,
                 type_,category,city,country,venue,event_web,googlePlaceUrl,mail_,
                 Speakerlist,on_off])

print('Done')
```
