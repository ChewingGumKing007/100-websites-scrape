```python
url='https://www.ocr.org.uk/qualifications/professional-development/events/?types=Conference&formats=Online%20webinar&locations=Online%20webinar&currentpage=1'

driver.get(url)
time.sleep(5)
soup = bs(driver.page_source,'lxml')

cards=[]

while True:
    soup = bs(driver.page_source,'lxml')
    card=soup.find_all('li',{'class':'course-entry'})
    cards.append(card)
    try:
        dim=soup.find('ul',{'class':'pagination'}).find_all('li')[-1].find('a')['href']
        driver.get(dim) 
    except:
        break
cards=list(itertools.chain.from_iterable(cards))  
len(cards)

sanp=[]
for m in cards:
# for k in range(1,22):
#     driver.switch_to.window(driver.window_handles[k])
#     time.sleep(1)
#     soupn=bs(driver.page_source,'lxml')
    tit=' '.join(m.h2.text.split())
    web=m.find('a')['href']
    link='https://www.ocr.org.uk'+web
    info=' '.join(m.p.text.split())
    timuu=' '.join(m.find('p',{'class':'start-date'}).get_text(separator='{').split())
    dat=timuu.split('{')[1]
    ti=timuu.split('{')[2]
    
    sanp.append([tit,link,info,dat,ti])
```

```python
ocr=[]
for k in sanp:
    link=k[1]
    driver.get(link)
    time.sleep(6)
    soupn=bs(driver.page_source,'lxml')
    
    titl=' '.join(soupn.find('li',{'class':'last-child'}).text.split())
    shim=soupn.find('div',{'class':'course-description'}).find_all('p')
    risj=[]
    for a in shim:
        if ':' not in str(a):
            risj.append(a)
        else:
            break
    info=' '.join([' '.join(b.text.split()) for b in risj])
    check=' '.join(soupn.find('div',{'class':'key-info-box'}).find('table').tbody.text.split())
    if 'arious locations' in check:
        tab=soupn.find('table',{'class':'ocr-new-table ocr-new-datetable'}).tbody.find_all('tr')
        loc=[' '.join(d.find('td').text.split()) for d in tab]
        ti=[' '.join(e.find('td',{'class':'desktop-time'}).text.split()) for e in tab]
        da=[' '.join(f.find('span',{'class':'desktop-date'}).text.split()) for f in tab]
        pri=[' '.join(g.find_all('td')[3].text.split()) for g in tab]
    else:
        shi=soupn.find('table',{'class':'key-info-table'}).find_all('tr')
        loc=[' '.join(soupn.find('table',{'class':'key-info-table'}).find('span',{'class':'course-loc'}).text.split())]
        da=[' '.join(soupn.find('table',{'class':'key-info-table'}).find('span',{'class':'course-date'}).text.split())]
        pri=[' '.join(soupn.find('table',{'class':'key-info-table'}).find('span',{'class':'course-fee'}).text.split())]
        ti=[' '.join(shi[1].find_all('td')[1].text.split()).replace(dat[0],'').strip()]
    for idx,h in enumerate(loc):
        title=titl
        event_info=info
        loca=h
        dat=da[idx]
        for m in ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']:
            if m in dat:
                dat=dat.replace(m,'')
        dat=dat.replace('–','-').replace('|','').replace('to','-').replace('TO','-').strip().replace('Oc-ber','October').replace('OC-BER','OCTOBER')
        if '(' in dat:
            dat=dat.split('(')[0].strip()
        if dat=='':
            start_date=end_date=''
        else:
            mad=dat
            #('September 8, 2022 I The Ritz-Carlton, Beijing',
        if re.search('[A-Sa-z]{3,9}\s?\d{1,2}\s+\d{4}\s?[A-Za-z]+\s?[A-Za-z]+\s?[A-Za-z]+',mad.replace(',','')):
            #print('juice')
            tad=re.search('[A-Sa-z]{3,9}\s?\d{1,2}\s+\d{4}',mad.replace(',','')).group()
            mad=tad
            st=mad.strip().replace(',','').replace('nd','').replace('st','').replace('th','').replace('rd','').replace('ugu','ugust')
            en=mad.strip().replace(',','').replace('nd','').replace('st','').replace('th','').replace('rd','').replace('ugu','ugust')

        elif '-' in mad:
            st=mad.split('-')[0].strip().replace(',','').replace('nd','').replace('st','').replace('th','').replace('rd','').replace('ugu','ugust')
            en=mad.split('-')[1].strip().replace(',','').replace('nd','').replace('st','').replace('th','').replace('rd','').replace('ugu','ugust')

        elif 'and'in mad:
            st=mad.split('and')[0].strip().replace(',','').replace('nd','').replace('st','').replace('th','').replace('rd','').replace('ugu','ugust')
            en=mad.split('and')[1].strip().replace(',','').replace('nd','').replace('st','').replace('th','').replace('rd','').replace('ugu','ugust')

        elif '&' in mad:
            st=mad.split('&')[0].strip().replace(',','').replace('nd','').replace('st','').replace('th','').replace('rd','').replace('ugu','ugust')
            en=mad.split('&')[1].strip().replace(',','').replace('nd','').replace('st','').replace('th','').replace('rd','').replace('ugu','ugust')

        else:
            #print(mad)
            st=mad.strip().replace(',','').replace('nd','').replace('st','').replace('th','').replace('rd','').replace('ugu','ugust')
            en=mad.strip().replace(',','').replace('nd','').replace('st','').replace('th','').replace('rd','').replace('ugu','ugust')


        if any(c.isalpha() for c in st)==False:
        #      print('leg')
            pa=re.search('[A-Sa-z]+\W+(\d{4})',en)#, maxsplit=0)
            sapa=pa.group()
            start=st+' '+sapa
            end=en
        #                        02        April      2022
        elif re.search('\d{1,2}\s+[A-Sa-z]{3,9}\s\d{4}',st):
         #   print('awujale')
            start=st
            end=en

        #*                        April       02      2022
        elif re.search('[A-Sa-z]{3,9}\s\d{1,2}\s+\d{4}',st):
        #       print('ala')
            pa=re.search('(\d{4})',st).group()
            sa=re.search('([A-Sa-z]{3,9})',st).group()
            ta=re.search('(\d{1,2})',st).group()
            start=ta+' '+sa+' '+pa
            #
            ba=re.search('(\d{4})',en).group()
            ca=re.search('([A-Sa-z]{3,9})',en).group()
            da=re.search('(\d{1,2})',en).group()
            end=da+' '+ca+' '+ba     

        #                      02     2022
        elif re.search('\d{1,2}\s+\d{4}',en):
        #      print('kum')
            pa=re.search('(\d{4})',en).group()#, maxsplit=0)
            sa=re.search('([A-Sa-z]{3,9})',st).group()
            ta=re.search('(\d{1,2})',st).group()

            start=ta+' '+sa+' '+pa
            #
            ba=re.search('(\d{4})',en).group()
            ca=re.search('([A-Sa-z]{3,9})',st).group()
            da=re.search('(\d{1,2})',en).group()
            end=da+' '+ca+' '+ba

        #                      02         April
        elif re.search('\d{1,2}\s+[A-Sa-z]{3,9}',st):
        #        print('is')
            pa=re.search('(\d{4})',en)#, maxsplit=0)
            sapa=pa.group()
            start=st+' '+sapa
            end=en

        #*                      April           02
        elif re.search('[A-Sa-z]{3,9}\s+\d{1,2}',st):
        #      print('bad')
            pa=re.search('(\d{4})',en).group()#, maxsplit=0)
            sa=re.search('([A-Sa-z]{3,9})',st).group()
            ta=re.search('(\d{1,2})',st).group()

            start=ta+' '+sa+' '+pa
            #
            ba=re.search('(\d{4})',en).group()
            ca=re.search('([A-Sa-z]{3,9})',en).group()
            da=re.search('(\d{1,2})',en).group()
            end=da+' '+ca+' '+ba

        else:
        #     print('shik')
            start=end=''
        if start=='':
            start_date=end_date=''
        else:
            pick=[start, end]
            try:
                spl_dt_obj = [datetime.strptime(v.strip(), '%d %B %Y') for v in pick]
            except:
                spl_dt_obj = [datetime.strptime(v.strip(), '%d %b %Y') for v in pick]
            date_= [z.strftime('%Y-%m-%d') for z in spl_dt_obj]
            start_date=date_[0]
            end_date=date_[1]   
            
        price=pri[idx]    
        tim=ti[idx]
        stt,ett=[k.strip() for k in tim.replace('–','-').upper().split('-')]
        z=''
        time_st={   
            'type':'general',
            'Start_time':stt,
            'end_time':ett,
            'timezone':z,
            'days':'all'
        }
        
        time_=json.dumps([time_st],ensure_ascii=False)
        
        if price=='Free'or price=='free':
            ty='free'
            cu=''
            am=''
        elif re.search('\$?\£?\W*\d+',price):
            ty='paid'
            cu=re.search('\$?\£?',price).group()
            am=re.search('\d+',price).group()
            
        tic={
            'type':ty,
            'price':am,
            'currency':cu
        }
        
        ticket_list=json.dumps([tic],ensure_ascii=False)
        
        def gu(luc):
            google_url_for_location="https://www.google.com/search?q="+luc+"&oq="+luc+"&num=1"
            time.sleep(randint(0,3))
            driver.get(google_url_for_location)
            time.sleep(4)
            try:
                google_map_url=driver.find_element_by_id('lu_map').click()
            except:
                try:
                    google_map_url=driver.find_element_by_class_name('Xm7sWb').click()
                except:
                    google_map_url=driver.find_element_by_class_name('Lx2b0d').click()
            time.sleep(1)
            google_map_url=driver.current_url
    #                 print(google_map_url)
            return(google_map_url)
            ######################################
        def get_google_map_url(location):
            try:
                return(gu(location))
            except:
                sha=location.split(',')
                try:
                    gu(sha[-3])
                except:
                    try:
                        gu(sha[-2])
                    except:
                        try:
                            gu(sha[-1])
                        except Exception as e:
                            print(location, "; url didn't go through")
                            print(e)
                            return("")


        def countr(locale):
            try:
                google_url_for_location="https://www.google.com/search?q="+locale+"&oq="+locale+"&num=1"
                time.sleep(randint(0,3))
                driver.get(google_url_for_location)
                time.sleep(4)
                try:
                    google_map_url=driver.find_element_by_id('lu_map').click()
                except:
                    try:
                        google_map_url=driver.find_element_by_class_name('Xm7sWb').click()
                    except:
                        google_map_url=driver.find_element_by_class_name('Lx2b0d').click()
                time.sleep(1)
                try:
                    country=driver.find_element_by_class_name('x3AX1-LfntMc-header-title-VdSJob').text
                except:
                    country=driver.find_element_by_class_name('bwoZTb').text
    #                 print(country)
                return(country)
            except Exception as e:
                print(locale, ";country didn't go through")
                print(e)
                return("")

        #ticket_list=''
        org_name='Oxford Cambridge and RSA (OCR)'
        org_web='https://www.ocr.org.uk/'
        org_pro='OCR is a leading UK awarding body. We provide qualifications which engage people of all ages and abilities at school, college, in work or through part-time learning programmes. Our general and vocational qualifications equip students with the knowledge and skills they need for their future, helping them achieve their full potential.'
        logo=''    
        sponsor=''
        agendalist=''
        type_=''
        
        Speakerlist=''
        category=''
        mail_=json.dumps(['professionaldevelopment@ocr.org.uk'])
        event_web=link
        loc=loca
        if 'nline' in loc or 'ff Campus' in loc or loc=='' or loc==' ' or loc=='-' or loc=='Virtual' or 'nspecified' in loc or 'irtual' in loc:
            on_off='1'
            city=''
            venue=''
            country=''
            googlePlaceUrl=''
        else:
            on_off='0'
            if ',' in loc:
                city=loc.split(',')[-2]+', '+loc.split(',')[-1].strip()
                country=loc.split(',')[-1].strip()
            else:
                city=loc
                country=countr(city)
            venue=loc
            googlePlaceUrl=get_google_map_url(venue)
        
        
        ocr.append([link,title,start_date,end_date,time_,event_info,ticket_list,
                     org_pro,org_name,org_web,logo,sponsor,agendalist,
                     type_,category,city,country,venue,event_web,googlePlaceUrl,mail_,
                     Speakerlist,on_off])
print('Done')
```
