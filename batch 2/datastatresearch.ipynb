```python
url = 'https://datastatresearch.org/courses/'
```

```python
driver.get(url)
```

```python
soup = bs(driver.page_source,'lxml')
```

```python
table = soup.find('table',{'id':'myTable'})
```

```python
columns = [i.text.replace('\t','').replace('\n','') for i in table.thead.find_all('th')]
```

```python
body = table.tbody
```

```python
tr = body.find_all('tr')
```

```python
statre=[]
regex = '[0-9]+'
for j in tr:
    td = [k.text.strip() for k in j.find_all('td')[1:]]
    tit,duration,amount = td[:-1]
    matches = re.finditer(regex, amount, re.MULTILINE)
    both_amount = []
    for matchNum, match in enumerate(matches, start=1):
        both_amount.append(match.group())
#     sia=both_amount[1]
    link= j.find('a')['href']
    #each_data = [title,sia,link]
    driver.get(link)
    #data.append(each_data)
    soupn = bs(driver.page_source,'lxml')
    maio=' '.join(soupn.find('div',{'id':'logo_row'}).find('div',{'class':'col-md-8'}).text.split()).split()
    for abd in maio:
        if '@' in abd:
            mass=abd
    mai=soupn.find('div',{'class':'caption'}).find_all('p',{'class':'MsoNormal'})
    
    for idx,a in enumerate(mai):
        if a.text=='Introduction':
            info=' '.join(mai[idx+1].text.split())
        elif 'INTRODUCTION' in a.text:
            info=' '.join(mai[idx].text.replace('INTRODUCTION','').strip().split())
    #     else:
    #         info=''
    tab=soupn.find('div',attrs={'class':'thumbnail','id':'desc'}).find('table')
    shi=tab.tbody.find_all('tr')
    we=soupn.find('div',attrs={'class':'thumbnail','id':'desc'}).find('p').text
    # wema=[v['href'] for v in we]
    sami=list(a.text.replace('\n\n','\n').replace(' ','').replace('.\n','').strip().split('\n\n\n')[0].split('\n') for a in shi)
    if 'USD' in we:
        curr='$'
    #*************
    assi=[]
    for a in sami:
        spl_dt_obj = [datetime.strptime(v, '%d/%m/%Y') for v in a]
        ate_= [z.strftime('%Y-%m-%d') for z in spl_dt_obj]
        assi.append(ate_)
    #******start********
   
    for m in assi:
        start_date=m[0]
        end_date=m[1]
        title=soupn.find('div',{'class':'container-fluid'}).find('div',{'class':'col-md-12'}).find('h3').text.replace('»','').replace('\n','').strip()
        try:
            if mass=='':
                mail_=''
            else:
                mail_=json.dumps([mass])
        except:
            mail_=''
        if 'GBP' not in we or 'USD' not in we or 'EUR' not in we:
            ticket_list=''
        else:
            if 'GBP' in we:
                sia=we.split('GBP')[-1].replace(',','').strip()
                if int(sia)>0:
                    ty='paid'
                    am=sia
                    cu='£'

                else:
                    ty='free'
                    am=''
                    cu=''
            elif 'USD' in we:
                sia=we.split('USD')[-1].replace(',','').strip()
                if int(sia)>0:
                    ty='paid'
                    am=sia
                    cu='$'
                else:
                    ty='free'
                    am=''
                    cu=''
            elif 'EUR' in we:
                sia=we.split('EUR')[-1].replace(',','').strip()
                if int(sia)>0:
                    ty='paid'
                    am=sia
                    cu='€'
                else:
                    ty='free'
                    am=''
                    cu=''
            else:
                ty=''
                am=''
                cu=''
            if ty=='':
                ticket_list=''
            else:
                ticketlist={'type': ty,
                             'price':am,
                             'currency': cu}
                ticket_list=json.dumps([ticketlist])
        event_info=info

        org_name='Datastat Research'
        org_web='https://datastatresearch.org/'
        on_off='1'
        if on_off=='1':
            city=''
            country=''
            venue=''
            googlePlaceUrl=''
        org_pro=''
        logo=''
        sponsors=''
        agendalist=''
        type_=''
        category=''
        event_web=link
        time_=''
        Speakerlist=''
        statre.append([link,title,start_date,end_date,time_,event_info,ticket_list,
                     org_pro,org_name,org_web,logo,sponsors,agendalist,
                     type_,category,city,country,venue,event_web,googlePlaceUrl,mail_,
                     Speakerlist,on_off])
print('Done')
```
